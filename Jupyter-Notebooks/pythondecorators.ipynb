```python
def dec1(func1):
    x = str(input('enter the operation : '))
    def exex():
      if x == 'yes':
        print('executing function')
        func1()
        print('completed')
      else:
       print('no instructions')
    return exex


@dec1
def arithimatic():
  a = 10
  b = 20
  task = str(input('enter task to perform : '))
  if task == 'addition':
    print(a + b)
  else:
    print(a - b)



arithimatic()
```

```python
def gen(n):
  fact = 1
  for i in range(1,n):
    fact = i * fact
    yield fact

g = gen(6)
for _ in range(5):
  print(g.__next__())

for itr in gen(6):
  print('factorial : ',itr)
```
