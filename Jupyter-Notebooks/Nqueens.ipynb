```python
possible_solutions = []

class N_Queens:

  def isSafe(self,mat, row, col):
    for i in range(row):
      if mat[i][col] == 'Q':
        return False
    i, j = row, col
    while i >= 0 and j >= 0:
        if mat[i][j] == 'Q':
            return False
        i = i - 1
        j = j - 1

    i, j = row, col
    while i >= 0 and j < len(mat):
        if mat[i][j] == 'Q':
            return False
        i = i - 1
        j = j + 1
 
    return True

  def nQueen(self,mat,row):

    if row == len(mat):
      return self.printSolution(mat)
    for i in range(len(mat)):
 
        if self.isSafe(mat, row, i):
            mat[row][i] = 'Q'
           
            self.nQueen(mat, row + 1)
            
            mat[row][i] = '-'

  def printSolution(self,mat):
    for r in mat:
      possible_solutions.append(str(r).replace(',', '').replace('\'', ''))


if __name__ == '__main__':
 
    q = N_Queens()
    N = int(input('enter your choice : '))

    mat = [['-' for x in range(N)] for y in range(N)]
 
    q.nQueen(mat, 0)
    solution_ = ["".join(i) for i in possible_solutions]
    print(len(solution_)/N)
```
