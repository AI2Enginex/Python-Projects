```python
import json
import nltk
nltk.download('stopwords')
import re
import pandas as pd
from nltk.corpus import stopwords
```

```python
json_file = open("/content/drive/MyDrive/Colab Notebooks/quotes.json","r",encoding="utf-8")
x = json_file.read()

final_data = json.loads(x)
```

```python
import numpy as np
category = []
quote = []
for hash_map in final_data:
    quote.append(hash_map["Quote"])
    category.append(hash_map["Category"])
```

```python
quote_category = pd.DataFrame({
    'quote' :  quote,
    'category' : category
})
```

```python
quote_category = quote_category.drop_duplicates(keep = "first" , subset='quote')
```

```python
quote_category.head(30)
```

```python
quote_category.isnull().sum()
```

```python
quote_category['quote'] = quote_category['quote'].str.lower()
```

```python
quote_category.head()
```

```python
code_category = list()

for message in quote_category['quote'].str.lower():
    word = re.sub('[^a-zA-Z]',' ' , message)
    review = word.split()
    stop_words = [word for word in review if word not in stopwords.words('english')]
    words = ' '.join(stop_words)
    code_category.append(words.split())
```

```python
code_category = [x for i in code_category for x in i]
```

```python
five_points = pd.read_csv("/content/drive/MyDrive/Colab Notebooks/file_content.csv")
```

```python
clean_text = []
for message in five_points['file_contents'].str.lower():
    word = re.sub('[^a-zA-Z]',' ' , message)
    review = word.split()
    stop_words = [word for word in review if word not in stopwords.words('english')]
    words = ' '.join(stop_words)
    clean_text.append(words.split())
```

```python
clean_text = [word for text in clean_text for word in text]
```

```python
print(len(code_category) , len(clean_text))
```

```python
common_words = list()

for word in code_category:
  if word in clean_text:
    common_words.append(word)
```
