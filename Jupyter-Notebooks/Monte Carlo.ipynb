```python
import random
import numpy as np


class Monte_Carlo:

  def __init__(self):
    self.res = 0
    self.org_deck = [
                  'AS', '2S', '3S', '4S', '5S', '6S', '7S', '8S', '9S', '10S', 'JS', 'QS', 'KS',
                  'AD', '2D', '3D', '4D', '5D', '6D', '7D', '8D', '9D', '10D', 'JD', 'QD', 'KD',
                  'AC', '2C', '3C', '4C', '5C', '6C', '7C', '8C', '9C', '10C', 'JC', 'QC', 'KC',
                  'AH', '2H', '3H', '4H', '5H', '6H', '7H', '8H', '9H', '10H', 'JH', 'QH', 'KH',
                ]

  def king_verification(self,x_):
     for i in range(len(x_)-1):
       if x_[i][0] == 'K' and x_[i+1][0] == 'K':
         return True
       else:
         pass

  def montecarlo(self):
    for _ in range(len(self.org_deck)-1):
      deck_ = np.copy(self.org_deck)
      random.shuffle(deck_)
      if self.king_verification(deck_):
        self.res = self.res + 1
    print(self.res/len(self.org_deck))
      
m = Monte_Carlo()

print(f"probability of two kings together ")
m.montecarlo()
```
